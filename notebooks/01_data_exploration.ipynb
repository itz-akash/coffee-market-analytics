```python
import pandas as pd
```

```python
coffee = pd.read_csv("../data/raw/psd_coffee.csv")

population = pd.read_csv(
    "../data/raw/API_SP.POP.TOTL_DS2_en_csv_v2_451027.csv",
    skiprows=4
)

countries = pd.read_csv(
    "../data/raw/countries-codes.csv",
    sep=";"
)
```

```python
print("Coffee:", coffee.shape)
print("Population:", population.shape)
print("Countries:", countries.shape)

coffee.head()
```

```python
print("Coffee Columns")
print(coffee.columns.tolist())

print("\nPopulation Columns")
print(population.columns.tolist())

print("\nCountries Columns")
print(countries.columns.tolist())
```

```python
countries.info()
```

```python
# For missing values

print("Coffee Missing Values")
print(coffee.isnull().sum())

print("\nPopulation Missing Values")
print(population.isnull().sum())

print("\nCountries Missing Values")
print(countries.isnull().sum())
```

```python
# Checking duplicates

print("Coffee duplicates:", coffee.duplicated().sum())
print("Population duplicates:", population.duplicated().sum())
print("Countries duplicates:", countries.duplicated().sum())
```

```python
# =========================
# DATA CLEANING
# =========================

# Coffee
coffee_clean = coffee.copy()

# Population
population_clean = population.copy()

# Drop completely empty columns
population_clean.drop(columns=["Unnamed: 70", "2025"], inplace=True)

# Countries
countries_clean = countries.copy()

# Keep only useful columns
countries_clean = countries_clean[
    ["ISO2 CODE", "ISO3 CODE", "LABEL EN"]
]

print("Cleaning completed!")

print(coffee_clean.shape)
print(population_clean.shape)
print(countries_clean.shape)
```

```python
# Convert population data from wide to long format

population_long = population_clean.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="Population"
)

# Convert Year to integer
population_long["Year"] = population_long["Year"].astype(int)

print(population_long.shape)
population_long.head()
```

```python
# =========================
# TRANSFORM POPULATION DATA
# =========================

population_long = population_clean.melt(
    id_vars=[
        "Country Name",
        "Country Code",
        "Indicator Name",
        "Indicator Code"
    ],
    var_name="Year",
    value_name="Population"
)

# Convert Year to integer
population_long["Year"] = population_long["Year"].astype(int)

print("Population Long Shape:", population_long.shape)

population_long.head()
```

```python
print(population_long.isnull().sum())
```

```python
population_long = population_long.dropna(subset=["Population"])

print(population_long.shape)
```

```python
print(
    population_long["Year"].min(),
    population_long["Year"].max()
)
```

```python
# =========================
# MERGE COFFEE + POPULATION
# =========================

coffee_population = coffee_clean.merge(
    population_long,
    left_on=["Country_Code", "Market_Year"],
    right_on=["Country Code", "Year"],
    how="left"
)

print("Merged Shape:", coffee_population.shape)

coffee_population.head()
```

```python
print("Missing Population after merge:")
print(coffee_population["Population"].isnull().sum())
```

```python
# =========================
# ADD ISO3 CODE TO COFFEE
# =========================

coffee_iso = coffee_clean.merge(
    countries_clean,
    left_on="Country_Code",
    right_on="ISO2 CODE",
    how="left"
)

print(coffee_iso.shape)

coffee_iso.head()
```

```python
print(coffee_iso[["Country_Code", "ISO2 CODE", "ISO3 CODE"]].head(10))
```

```python
left_on = ["ISO3 CODE", "Market_Year"]
right_on = ["Country Code", "Year"]
```

```python
# =========================
# FINAL MERGE
# =========================

final_df = coffee_iso.merge(
    population_long,
    left_on=["ISO3 CODE", "Market_Year"],
    right_on=["Country Code", "Year"],
    how="left"
)

print("Final Shape:", final_df.shape)

print("\nMissing Population:")
print(final_df["Population"].isnull().sum())

final_df.head()
```

```python
matched = final_df["Population"].notnull().sum()

print(f"Matched Records: {matched}")
print(f"Total Records: {len(final_df)}")
print(f"Match Rate: {matched/len(final_df)*100:.2f}%")
```

```python
# Countries that did not match with population

unmatched = final_df[final_df["Population"].isna()][
    ["Country_Name", "Country_Code", "ISO3 CODE"]
].drop_duplicates()

print("Number of unmatched countries:", len(unmatched))
unmatched.sort_values("Country_Name").head(50)
```

```python
country_name_mapping = {
    "Cote d'Ivoire": "Côte d'Ivoire",
    "Congo (Brazzaville)": "Congo",
    "Congo (Kinshasa)": "Democratic Republic of the Congo",
    "Korea, South": "South Korea",
    "Laos": "Lao People's Democratic Republic",
    "Russia": "Russian Federation",
    "Syria": "Syrian Arab Republic",
    "Turkey": "Türkiye"
}
```

```python
# Normalize country names before joining

country_name_mapping = {
    "Congo (Brazzaville)": "Congo",
    "Congo (Kinshasa)": "Democratic Republic of the Congo",
    "Cote d'Ivoire": "Côte d'Ivoire",
    "Iran": "Iran (Islamic Republic of)",
    "Korea, South": "Republic of Korea",
    "Laos": "Lao People's Democratic Republic",
    "North Macedonia": "Macedonia",
    "Russia": "Russian Federation",
    "Taiwan": "Taiwan",
    "Tanzania": "United Republic of Tanzania",
    "Venezuela": "Venezuela (Bolivarian Republic of)",
    "Vietnam": "Viet Nam",
    "Yemen (Sanaa)": "Yemen"
}

coffee_clean["Country_Name_Normalized"] = (
    coffee_clean["Country_Name"]
    .replace(country_name_mapping)
)
```

```python
coffee_country = coffee_clean.merge(
    countries_clean,
    left_on="Country_Name_Normalized",
    right_on="LABEL EN",
    how="left"
)

print(coffee_country.shape)

coffee_country.head()
```

```python
population_long["Country Name"] = population_long["Country Name"].replace({
    "Congo, Dem. Rep.": "Congo (Kinshasa)",
    "Congo, Rep.": "Congo (Brazzaville)",
    "Iran, Islamic Rep.": "Iran",
    "Korea, Rep.": "Korea, South",
    "Lao PDR": "Laos",
    "Viet Nam": "Vietnam",
    "Venezuela, RB": "Venezuela",
    "Yemen, Rep.": "Yemen",
    "Tanzania, United Rep.": "Tanzania"
})
```

```python
final_df = coffee_country.merge(
    population_long,
    left_on=["ISO3 CODE", "Market_Year"],
    right_on=["Country Code", "Year"],
    how="left"
)

matched = final_df["Population"].notnull().sum()

print("Matched:", matched)
print("Missing:", final_df["Population"].isnull().sum())
print(f"Match Rate: {matched/len(final_df)*100:.2f}%")
```

```python
analysis_df = final_df[[
    "Country_Name",
    "Country_Code",
    "ISO3 CODE",
    "Market_Year",
    "Attribute_Description",
    "Value",
    "Population"
]].copy()

analysis_df.head()
```
